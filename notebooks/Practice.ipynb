```python
from scipy.integrate import odeint
import pandas as pd
import numpy as np

diff = pd.read_csv('Book1.csv')['Enthalpy deficit'].to_numpy()
n = len(diff)

z = np.linspace(0, n+1, n+1)
T_0 = 314
T_arr = [T_0]

def fun(z, T, diff):
    
    return 0.1475*diff/2000
    
for i in range(n):

    z_span = [z[i], z[i+1]]
    T = odeint(fun, T_arr[i], z_span, args=(diff[i],)).T[-1][-1]
    print(T)
    T_arr.append(T)
    

print(T_arr)
```
